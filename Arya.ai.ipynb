```python
column_names = df.columns[:-1]
```

```python
# checking for unique values in  each columns
for i in column_names:
  print(df[i].unique())
```

```python
# Number of zeros
for i in column_names: 
  print(i, (df[i] == 0).sum())
```

```python
# zeros by percentage
for i in column_names: 
  print(i, round(((df[i] == 0).sum() / len(df[i]))* 100, 2))
```

```python
for i in column_names:
  plt.title(i)
  sns.boxplot(x= df["Y"], y = df[i])
  plt.show()
```

```python
for i in column_names:
  plt.title(i)
  plt.hist(df[i])
  plt.xlabel("Values")
  plt.show()
  print("skewness is : ", df[[i]].skew())
  
```

```python

from scipy import stats
import pylab 
for i in column_names:
  stats.probplot(df[i], dist= "norm", plot=pylab)
  plt.title(i)
  pylab.show()
```

```python
#Applying log trasnformation
for i in column_names:
  plt.hist(np.log(df[i] + 1))
  plt.show()
  print("skewness after tansormation: ", (np.log(df[column_names[0]] + 1)).skew())
  print("Previous skewness :", df[i].skew())

```
