```python
df = pd.read_csv("training_set.csv")
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
x = df.drop(columns = ["Unnamed: 0", "Y"])
```

```python
x.head()
```

```python
y = df["Y"]
```

```python
y
```

```python
X_train, X_test, y_train , y_test = train_test_split(x, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# random forest base model
from sklearn.ensemble import RandomForestClassifier
r_c = RandomForestClassifier()
```

```python
r_c.fit(X_train, y_train)
```

```python
y_pred = r_c.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

# predicting test data

```python
# Predicting test data
test = pd.read_csv("test_set.csv")
test.head()
```

```python
x = test.drop(labels="Unnamed: 0", axis=1)
x.shape
```

```python
y_pred_test = r_c.predict(x)
y_pred_test = (y_pred_test > 0.5)
y_pred_test
y_pred_int = y_pred_test.astype("int")
```

```python
y_pred_int
```

```python
test["y_pred"] = y_pred_int
test.shape

```
